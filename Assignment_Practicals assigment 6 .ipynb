```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# 1. Create a 2D NumPy array and calculate row-wise sum
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = arr.sum(axis=1)
print("Row sums:", row_sum)

# 2. Find mean of a specific column in Pandas
df = pd.DataFrame({"Name": ["A", "B", "C"], "Marks": [80, 90, 75]})
print("Mean of Marks:", df["Marks"].mean())

# 3. Scatter plot using Matplotlib
plt.scatter([1,2,3,4], [10,20,25,30])
plt.title("Scatter Plot")
plt.show()

# 4. Correlation matrix heatmap using Seaborn
data = sns.load_dataset("iris")
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()

# 5. Bar plot using Plotly
fig = px.bar(df, x="Name", y="Marks", title="Bar Plot")
fig.show()

# 6. Add new column in DataFrame
df["Result"] = ["Pass" if x > 80 else "Fail" for x in df["Marks"]]
print(df)

# 7. Element-wise multiplication of two NumPy arrays
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Element-wise multiplication:", a*b)

# 8. Line plot with multiple lines
x = [1,2,3,4]
plt.plot(x, [1,4,9,16], label="y=x^2")
plt.plot(x, [1,2,3,4], label="y=x")
plt.legend()
plt.show()

# 9. Filter DataFrame rows where value > threshold
filtered = df[df["Marks"] > 80]
print("Filtered Data:\n", filtered)

# 10. Histogram using Seaborn
sns.histplot(df["Marks"], bins=5, kde=True)
plt.show()

# 11. Matrix multiplication using NumPy
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print("Matrix multiplication:\n", np.dot(A,B))

# 12. Load CSV and display first 5 rows
# df_csv = pd.read_csv("file.csv")
# print(df_csv.head())

# 13. 3D scatter plot using Plotly
fig = px.scatter_3d(data, x="sepal_length", y="sepal_width", z="petal_length", color="species")
fig.show()
```
